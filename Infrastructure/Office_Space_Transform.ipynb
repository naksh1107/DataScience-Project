# Feature Engineering 

```python

def get_area_name(api_key, place_name):
    url = f"https://maps.googleapis.com/maps/api/place/findplacefromtext/json"
    params = {
        'input': place_name,
        'inputtype': 'textquery',
        'fields': 'formatted_address',
        'key': api_key
    }
    response = requests.get(url,params=params)

    if response.status_code == 200:
        data = response.json()
        
        if 'candidates' in data and len(data['candidates']) > 0:
            result = data['candidates'][0]
            if 'formatted_address' in result:
                return result['formatted_address']
            else:
                return "Area name not found in the response."
        else:
            return "No result found in the response."
    else:
        return f"Error: HTTP {response.status_code}"
```

```python
Area_Name1=[]
Area_Name=[]
api_key = 'Enter Your API Key'
for i in range(len(df7)):
    place_name=df7.loc[i,"Location"]
    area_name = get_area_name(api_key, f"{place_name},Bangalore")
    Area_Name1.append(area_name.split(','))

for j in Area_Name1:
    newList=[area.strip() for area in j]
    if "Bengaluru" in newList:
        prev=''
        
        for i in newList:
            if i=="Bengaluru":
                Area_Name.append(prev)

            prev=i
    elif any(area.strip().startswith("Karnataka") for area in newList):
        prev=''
        for k in newList:
            if k.startswith("Karnataka"):
                Area_Name.append(prev)
            prev=k
len(Area_Name)
```

```python
def find_nearest_metro(location,Distance_from_station,metros):
    distanceList=[]
    distanceDict={}
    gmaps = googlemaps.Client(key='Enter Your API Key')
    place_coordinates=gmaps.geocode(location)
    
    lat_origin=place_coordinates[0]['geometry']['location']['lat']
    long_origin=place_coordinates[0]['geometry']['location']['lng']
    origin_location = str(lat_origin)+','+str(long_origin)  # Latitude, Longitude
    radius = 5000 # Search radius in meters
    type = "subway_station"
    places = gmaps.places_nearby(location=origin_location, radius=radius, type=type)
    for place in places["results"]:
        lat_destination=place['geometry']['location']['lat']
        long_destination=place['geometry']['location']['lng']
        destination_location=str(lat_destination)+','+str(long_destination)
        distance=gmaps.distance_matrix(origins=origin_location,destinations=destination_location)
        distance_main=distance["rows"][0]["elements"][0]["distance"]["value"]
        distanceList.append(distance_main)
        name=place['name']
        distanceDict[name]=distance_main
        
    distanceList.sort()
    value = [i for i in distanceDict if distanceDict[i]==distanceList[0]]
    Distance_from_station.append(distanceDict[value[0]])
    metros.append(value[0])
    return value[0]
    
```

```python
column_names=df7['Area Name'].tolist()
Nearby_metro=[]
Distance_from_station=[]
metros=[]
for i in range(len(column_names)):
    try:
        metro=find_nearest_metro(f'{column_names[i]} Bangalore',Distance_from_station,metros)
        Nearby_metro.append(metro)
    except:
        Nearby_metro.append("NA")
        Distance_from_station.append("NA")


```

```python
df7['Nearby Metro Station']=Nearby_metro
df7['Distance From Metro']=Distance_from_station
```
