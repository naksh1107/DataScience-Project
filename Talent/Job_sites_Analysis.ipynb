# Data Cleaning

```python
df1=df1[~(df1["Applicants"]=="Less than 10")]
df2=df2[~(df2["Applicants"]=="Less than 10")]
df3=df3[~(df3["Applicants"]=="Less than 10")]
df4=df4[~(df4["Applicants"]=="Less than 10")]
df5=df5[~(df5["Applicants"]=="Less than 10")]
df6=df6[~(df6["Applicants"]=="Less than 10")]
```

```python
df1=df1.dropna(subset=['Openings'])
df2=df2.dropna(subset=['Openings'])
df3=df3.dropna(subset=['Openings'])
df4=df4.dropna(subset=['Openings'])
df5=df5.dropna(subset=['Openings'])
df6=df6.dropna(subset=['Openings'])
```

```python
sum_applicants_frontend_blr=np.sum(df1["Applicants"].astype(int))
sum_applicants_backend_blr=np.sum(df2["Applicants"].astype(int))
sum_applicants_engineering_blr=np.sum(df3["Applicants"].astype(int))
sum_applicants_frontend_hyd=np.sum(df4["Applicants"].astype(int))
sum_applicants_backend_hyd=np.sum(df5["Applicants"].astype(int))
sum_applicants_engineering_hyd=np.sum(df6["Applicants"].astype(int))
sum_Openings_frontend_blr=np.sum(df1["Openings"].astype(int))
sum_Openings_backend_blr=np.sum(df2["Openings"].astype(int))
sum_Openings_engineering_blr=np.sum(df3["Openings"].astype(int))
sum_Openings_frontend_hyd=np.sum(df4["Openings"].astype(int))
sum_Openings_backend_hyd=np.sum(df5["Openings"].astype(int))
sum_Openings_engineering_hyd=np.sum(df6["Openings"].astype(int))
```

# Data Analysis

### Total Number of Applicants

```python
print("Total Frontend Applicants in Bangalore:",sum_applicants_frontend_blr)
print("Total Backend Applicants in Bangalore:",sum_applicants_backend_blr)
print("Total Engineering Manager Applicants in Bangalore:",sum_applicants_engineering_blr)
print("Total Frontend Applicants in Hyderabad:",sum_applicants_frontend_hyd)
print("Total Backend Applicants in Hyderabad:",sum_applicants_backend_hyd)
print("Total Engineering Manager Applicants in Hyderabad:",sum_applicants_engineering_hyd)
```

```python
DS_front_end_blr=sum_Openings_frontend_blr/sum_applicants_frontend_blr
DS_back_end_blr=sum_Openings_backend_blr/sum_applicants_backend_blr
DS_engman_blr=sum_Openings_engineering_blr/sum_applicants_engineering_blr
DS_front_end_hyd=sum_Openings_frontend_hyd/sum_applicants_frontend_hyd
DS_back_end_hyd=sum_Openings_backend_hyd/sum_applicants_backend_hyd
DS_engman_hyd=sum_Openings_engineering_hyd/sum_applicants_engineering_hyd
```

## Demand Supply Ratio

```python
print("Demand Supply ratio of frontend Engineering job in Bangalore:",DS_front_end_blr)
print("Demand Supply ratio of Backend Engineering job in Bangalore:",DS_back_end_blr)
print("Demand Supply ratio of Engineering Manager job in Bangalore:",DS_engman_blr)
print("Demand Supply ratio of Frontend Engineering job in Hyderabad:",DS_front_end_hyd)
print("Demand Supply ratio of Backend Engineering job in Hyderabad:",DS_back_end_hyd)
print("Demand Supply ratio of Engineering Manager job in Hyderabad:",DS_engman_hyd)
print('\n')

print("Average Demand Supply Ratio in Bangalore:",(DS_front_end_blr+DS_back_end_blr+DS_engman_blr)/3)
print("Average Demand Supply Ratio in Hyderabad:",(DS_front_end_hyd+DS_back_end_hyd+DS_engman_hyd)/3)
```

# CONCLUSION

<span style="font-size:1.2em;">
The number of applicants for job titles is higher in Bangalore compared to Hyderabad. However, the demand/supply ratio in Hyderabad is lower, indicating that it is more challenging to find jobs there. This disparity highlights several reasons why opening an office in Bangalore is a more strategic choice:

1. Abundant Talent Pool: The higher number of applicants in Bangalore means a larger and more diverse talent pool, making it easier to find qualified candidates for various roles.

2. Dynamic Job Market: The higher job competition in Hyderabad suggests a more saturated job market, whereas Bangalore presents more opportunities due to its robust job market.
